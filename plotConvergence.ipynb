```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plotthings(vals, labels):
    fig = plt.figure(figsize=(10,5), facecolor='white')
    ax = fig.add_subplot(111)
    
    # ax.set_xscale('log', base=2)

    plt.tight_layout()
    plt.xlabel("Iterations", fontsize=16)
    plt.ylabel("Training Loss", fontsize=16)
    plt.title("ResNet50 Training Loss, Skip-Reduce", fontsize=20)

    for i in range(len(labels)):
        plt.plot(vals[i], label=labels[i])

    plt.legend(fontsize=16) 
    plt.show()
```

```python
# save .csv files into traces/resnet50 folder

# files = ["RESNET50", "RESNET50_R25", "RESNET50_R50", "RESNET50_R75", "RESNET50_R90"] 
# labels = ["Baseline", "25% Random", "50% Random", "75% Random", "90% Random"]

files = ["RESNET50", "RESNET50_X1" ,"RESNET50_X2"] 
labels = ["Baseline", "Skipping 1 RS","Skipping 2 RS"]


# files = ["VGG16", "VGG16_R25", "VGG16_R50"] 
# labels = ["Baseline", "25% Random", "50% Random"]

# files = ["VGG16", "VGG16_X1", "VGG16_X1Y1"] 
# labels = ["Baseline", "Skipping 1 RS", "Skipping 1 RS and 1 AG"]







folder = "./trace/"

```

```python
vals = []
for f in files:
    path = folder+f + ".csv"
    d = pd.read_csv(path)
    l = d["Loss"].astype(float).to_numpy()
    vals.append(l)
```

```python
plotthings(vals, labels)
```
